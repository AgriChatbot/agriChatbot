```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from nltk.probability import FreqDist
import operator
```

```python
df_list=[]
for i in range(1,36):
    df = pd.read_csv("11_%d.csv"%(i))
    df_list.append(df)
    
```

```python
data = pd.DataFrame()
data = pd.concat(df_list)
```

```python
data
```

```python
list(data)
```

```python

stats_list=['Category','Crop','DistrictName','QueryType','Season','Sector']
#stats_list=['DistrictName']
```

```python
def calcstats(stat_name):
    elementList = data[stat_name].values
    newDict = FreqDist(elementList)
    sortedList=np.array(sorted(newDict.items(), key=operator.itemgetter(1)))
    
    nameList=[]
    numQueries=[]
    for ele in sortedList:
        nameList.append(ele[0])
        numQueries.append(ele[1])
        
    return [nameList,numQueries]
```

```python
PlotLists=[]
for stat_name in stats_list:
    PlotList=calcstats(stat_name)
    PlotLists.append(PlotList)
```

```python
#PlotLists
```

# Category 

```python
# Category
plt.bar(PlotLists[0][0],PlotLists[0][1])
plt.xticks(rotation=90)
plt.show()
```

```python
fig1, ax1 = plt.subplots()
#explode = (0.05,0.05,0.05,0.05)
explode = [0.05 for i in range(11)]
explode =tuple(explode)
ax1.pie(PlotLists[1][1][-12:-1], labels=PlotLists[1][0][-12:-1], autopct='%1.1f%%',
        shadow=False, pctdistance=0.85,startangle=90,explode = explode)

#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  

plt.tight_layout()
plt.show()
```
