```python
stateCode = 11
districtCode = 2
df=pd.read_csv('/home/soham/nlp_project/scraping/data/{}_{}.csv'.format(stateCode,districtCode))
```

```python
queryText = df['QueryText']
KCCAns = df['KCCAns']
crop = df['Crop']
```

```python
for q in queryText[:]:
#     print(q)
    qs = re.findall(r'\w+', q.lower())
    for w in qs:
        if w in crop_hindi_eng_dict.keys():
            print(w,crop_hindi_eng_dict[w],'Hindi-English\n')
        else:
            if w!=spell_correct_functions.correction(w):
                print(w, spell_correct_functions.correction(w),'\n')
    print('')
```

```python
for i in range(1,len(queryText)):
    q = queryText[i]
    qs = re.findall(r'\w+', q.lower())
    if len(qs)>0:
        qsent = ''
        for w in qs:
            qsent += spell_correct_functions.correction(w) + ' '
        df.at[i,'QueryText']=qsent
    else:
        df.drop(i,inplace=True)
        print('dropped', qs)
```
