```python
import pandas as pd
import spell_correct_functions
import re
from nltk.corpus import wordnet,stopwords
from nltk.probability import FreqDist
from nltk.corpus import words

stop_words = set(stopwords.words('english')) 
```

```python
list_nonEng_words = []
def remove_all_english_words_queryText(df):
    queryText = df['QueryText']
    for i in range(len(queryText)):
        q = queryText[i]
        if type(q)!=str:
            df.drop(i,inplace=True)
            continue
        qs = re.findall(r'\w+', q.lower())
        if len(qs)>0:
            qsent = ''
            for w in qs:
                if not wordnet.synsets(w) and w not in stop_words:
                    qsent += w + ' '
            if len(qsent)>0:
                df.at[i,'QueryText']=qsent
                list_nonEng_words.append(qsent)
            else:
                df.drop(i,inplace=True)
        else:
            df.drop(i,inplace=True)
#             print('dropped', qs)
#     df.to_csv('./NoEng/Eng_words_removed_{}_{}.csv'.format(stateCode,districtCode))
```

```python
stateCode = 11
for districtCode in range(31,36):
    df=pd.read_csv('/home/soham/nlp_project/scraping/data/{}_{}.csv'.format(stateCode,districtCode))
    remove_all_english_words_queryText(df)
```
